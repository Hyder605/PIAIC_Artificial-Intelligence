````python
def student_card(roll_no,name,fname,course):
    card=f"""
    Piaic student Identity card
    
    Roll no : {roll_no}
    Name : {name}
    Father Name : {fname}
    Course : {course}
    """
    
    return card

print(student_card(12,"M ALI",'Ali',"AI"))

print(student_card("M Ali","AI",12,"Ali")) # if we gave arg values randomly then it will show result randomly

print(student_card(name="M Ali",course="AI",roll_no=12,fname="Ali")) #providing key arguments will help to solve the issue of randomness.

````

````python
l=[12, "M Ali", "Ali", "AI"]

print(student_card(*l))
````

````python
d1={
    '12':"Roll noo",
    'M Ali':"name",
    'Ali':"fname",
    'AI':"course"
}

print(student_card(*d1))
````

# Lambda Function

````python
def add(n1,n2):
    return n1+n2

add(2,3)
````

````python
#lambda arg1,arg2:function body

a=lambda n1,n2: n1+n2 #lambda don't require fucnction name
a(2,3)  #calling variable not function
````

````python
data=[[1,"Z"],
     [2,"Y"],[3,"X"]]

sorted(data,key=lambda x:x[1])
#iteration..1st take the x,as first list and return x[1]which is Z and so on..and on the basis of x[1] it sorted the value
````

````python
sorted(data,key=lambda x:x[0])
````

````python
sorted(data,key=lambda x:x[0],reverse=True) #descending order
````

````python
def  my_gen(s,e):
    for i in range(s,e+1):
        yield i
        
abc=my_gen(1,20)        
````

````python
print(abc) #will not show values until u iterate
````

````python
print(next(abc))
````

````python
print(next(abc))
# iterating each line step by step
````

````python
print(next(abc)) # iterating each line step by step
print(next(abc))
print(next(abc))
````

````python
for i in my_gen(1,6):
    print(i)
````

# Recursive Function

calling itself in its own function

````python
def abc(x):
    print(x)
    if x==0:
        return 
    else:
        abc(x-1) #function calling itself again
        
        
abc(5)


#let x=5
#print=====5
#checking conditions x==0?(is 5==0)
#condition fail...move to else
#now abc(x-1)...i.e abc(5-1)
#abc(4)....this is function,which we made>abc(x)<,,so call it again
#print====4
#checking conditions x==0?(is 4==0)
#condition fail...move to else
#now abc(x-1)...i.e abc(4-1)
#abc(3)...again function ..so calling itself again
#.
#.
#.
#. so on  until x==0


````
